```python
#Question 01
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
c=np.array([(50,50),(50,100),(150,255),(150,150)])
t1=np.linspace(0,c[0,1],c[0,1]+1-0).astype('uint8')
# t2=np.linspace(c[0,1]+1,c[1,1],0).astype('uint8')
t3=np.linspace(c[1,1]+1,c[2,1],c[2,0]-c[1,0]).astype('uint8')
# t4=np.linspace(c[3,1]+1,c[2,1],0).astype('uint8')
t5=np.linspace(c[2,0]+1,c[2,1],c[2,1]-c[3,0]).astype('uint8')
print(len(t1))
print(len(t3))
print(len(t5))
transform=np.concatenate((t1,t3),axis=0).astype('uint8')
transform=np.concatenate((transform,t5),axis=0).astype('uint8')
print(len(transform))
fig,ax=plt.subplots()
ax.plot(transform)
ax.set_xlabel(r'Input, $(\mathbf{x})$')
ax.set_ylabel('output, $\mathrm{T}[f(\mathbf{x})]$')
ax.set_xlim(0,255)
ax.set_ylim(0,255)
ax.set_aspect('equal')
plt.show()
img_org= cv.imread(r"C:\Users\User\Desktop\simulation\-Image-Processing-and-Machine-Vision\emma_gray.jpg")
assert img_org is not None
fig,ax=plt.subplots(1,2,figsize=(18,6))
ax[0].imshow(img_org)
ax[0].set_title('original')
# ax.plot(transform)
ax[0].set_aspect('equal')
assert len(transform) ==256
img_transformed=cv.LUT(img_org,transform)
ax[1].imshow(img_transformed)
ax[1].set_title('transformed_image')
ax[1].set_aspect('equal')
assert len(transform) ==256
cv.namedWindow('Image', cv.WINDOW_AUTOSIZE)
cv.imshow('Image',img_org)
cv.waitKey(0)
cv.imshow('Image',img_transformed)
cv.waitKey(0)
cv.destroyAllWindows()
```
